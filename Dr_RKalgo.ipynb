```python
%%time

#noval algorithm

from itertools import chain, combinations

frequentPattern ={};
tmpDict = {};

# import pandas lib as pd
import pandas as pd
 
# read 2nd sheet of an excel file
dataframe2 = pd.read_excel('u5.xlsx')
 

columnNames=list(dataframe2.columns)
print("column:"+str(columnNames))


#give support count to genrate  frequent pattern
sp = 0.9
minSup = int(sp*len(columnNames))
print("minSup : "+str(minSup)) 

frequentPattern ={};
tmpDict = {};


def generate_frequent_patterns(subset):
  input="";
  noOfItems=0;
  global frequentPattern;
  global tmpDict;
  for s in subset:
    for t in range(len(s)):
      input=input+s[t];
      noOfItems=noOfItems+1;
    #print("set:"+input);
    if input in frequentPattern:
      tmpDict["count"]=frequentPattern.get(input)["count"]+1;
      tmpDict["noOfItems"]=frequentPattern.get(input)["noOfItems"];
      frequentPattern[input]=tmpDict;
    else:
      tmpDict["count"]=1;
      tmpDict["noOfItems"]=noOfItems;
      frequentPattern[input]=tmpDict;
    
    #print("no of items in set  :"+str(noOfItems));
    input="";
    noOfItems=0;
    tmpDict={};
    print("\n");


def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(1,len(s)+1))


input=[];
for index,row in dataframe2.iterrows():
  for col in columnNames:
    #first col must be index
    if col!=columnNames[0]:
      #print("row[col] : "+str(row[col]))
      if row[col] == 1.0:
        input.append(col)  

  subset=list(powerset(input))
  #print("subset:"+str(subset))
  input=[]
  generate_frequent_patterns(subset);
print("frequentPattern found : ")
for dictEle in frequentPattern.keys():
  if frequentPattern[dictEle]["count"]>=minSup:
    print(dictEle + "  count :" + str(frequentPattern[dictEle]["count"]))


```

```python
#noval using additional storage

%%time


from itertools import chain, combinations

frequentPattern ={};
tmpDict = {};
cacheOfPowerSet={};
# import pandas lib as pd
import pandas as pd
 
# read 2nd sheet of an excel file
dataframe2 = pd.read_excel('u5.xlsx')
 

columnNames=list(dataframe2.columns)
#print("column:"+str(columnNames))

sp = 0.9
minSup = int(sp*len(columnNames))
print("minSup : "+str(minSup)) 

frequentPattern ={};
tmpDict = {};


def generate_frequent_patterns(subset):
  input="";
  noOfItems=0;
  global frequentPattern;
  global tmpDict;
  for s in subset:
    for t in range(len(s)):
      input=input+s[t];
      noOfItems=noOfItems+1;
    #print("set:"+input);
    if input in frequentPattern.keys():
      tmpDict["count"]=frequentPattern.get(input)["count"]+1;
      tmpDict["noOfItems"]=frequentPattern.get(input)["noOfItems"];
      frequentPattern[input]=tmpDict;
    else:
      tmpDict["count"]=1;
      tmpDict["noOfItems"]=noOfItems;
      frequentPattern[input]=tmpDict;
    
    #print("no of items in set  :"+str(noOfItems));
    input="";
    noOfItems=0;
    tmpDict={};
    print("\n");


def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(1,len(s)+1))


input=[];
for index,row in dataframe2.iterrows():
  for col in columnNames:
    #first col must be index
    if col!=columnNames[0]:
      #print("row[col] : "+str(row[col]))
      if row[col] == 1.0:
        input.append(col)  

  s = ''.join(input);    
  if s in cacheOfPowerSet:
    pwrset=cacheOfPowerSet[s];
    #print("reading from cache")
  else:
    pwrset=list(powerset(input));
    cacheOfPowerSet[s]=pwrset;
    #print("storing from cache")


  input=[]
  generate_frequent_patterns(pwrset);
print("frequentPattern found :")

for dictEle in frequentPattern.keys():
  if frequentPattern[dictEle]["count"]>=minSup:
    print(dictEle + "  count :" + str(frequentPattern[dictEle]["count"]))



```
